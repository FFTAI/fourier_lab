```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
data_root = '/home/fourier/GRX_humanoid/'
real_data_path = data_root + '/real data/251217_GR3_57号左臂失控/12.17-57号日志/aurora/motor_data_0.1.csv'
real_data = pd.read_csv(real_data_path)

print(real_data.shape)
start_id = 0
end_id = real_data.shape[0]
time_strings = real_data.iloc[start_id:end_id, 0]
# time = [datetime.strptime(t[11:], '%H:%M:%S.%3f') for t in time_strings]
# time_seconds = [(t - time[0]).total_seconds() for t in time]
# print(time_seconds[:10])

time_dt = []
for t in time_strings:
    time_part = t[11:]  # 跳过日期部分
    if '.' in time_part:
        hms, ms = time_part.split('.')
        ms6 = ms.ljust(6, '0')  # 补齐到微秒位
        t_norm = f"{hms}.{ms}"
        dt = datetime.strptime(t_norm, '%H:%M:%S.%f')
    else:
        dt = datetime.strptime(time_part, '%H:%M:%S')
    time_dt.append(dt)

t0 = time_dt[0]
time_seconds = np.array([(dt - t0).total_seconds() for dt in time_dt], dtype=float)
time_seconds = np.round(time_seconds, 3)  # 保留到毫秒
print(time_seconds[:10])
time_str_mm = [dt.strftime("%M:%S.%f")[:-3] for dt in time_dt]
print(time_str_mm[:10])  # 打印前10个（分钟以后）
```

```python
joint_pos = real_data.iloc[start_id:end_id, 2:2+31].values
joint_vel = real_data.iloc[start_id:end_id, 2+31+1:2+31*2+1].values
joint_tor = real_data.iloc[start_id:end_id, 2+31*2+2:2+31*3+2].values
jPos_cmd = real_data.iloc[start_id:end_id, 2+31*3+3:2+31*4+3].values
jVel_cmd = real_data.iloc[start_id:end_id, 2+31*4+4:2+31*5+4].values
jTor_cmd = real_data.iloc[start_id:end_id, 2+31*5+5:2+31*6+5].values
```

```python
# joint pos
# lower joint
fig, axs = plt.subplots(4, 4, figsize=(40, 20), sharex=True)
for i, ax in enumerate(axs.flatten()):
    ax = axs.flatten()[i]
    if i < 15:
        ax.plot(time_dt, jPos_cmd[:, i], label=f'pos_cmd_{lower_joint_names[i]}')
        ax.plot(time_dt, joint_pos[:, i], label=f'pos_{lower_joint_names[i]}')
        ax.legend()
fig.suptitle('lower Joint Position')
plt.tight_layout()
```
